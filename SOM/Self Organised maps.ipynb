```python
from sklearn.datasets import load_iris 
```

```python
iris=load_iris()
```

```python
data=iris.data
features=iris.feature_names
target=iris.target
labels=iris.target_names
```

```python
#!conda install MiniSom
```

```python
from minisom import MiniSom
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#normalise using numpy
data=np.apply_along_axis(lambda x:x/np.linalg.norm(x),1,data)
```

```python
# instantiating
som=MiniSom(x=10,y=10,input_len=4,sigma=1.0,learning_rate=0.5)
```

```python
#randomly initialise weight
som.random_weights_init(data=data)
```

```python
# training som
som.train_random(data=data,num_iteration=100)
```

```python
data.shape
```

```python
#plotting the responses
plt.bone()
plt.pcolor(som.distance_map())
plt.colorbar()
plt.show()
```

```python
#using markers for visualising
markers=['o','s','D']
colors=['r','g','b']# red :setosa g: versicolor blue diamond verginica

for i,x in enumerate(data):
    #get the winning nodes
    w=som.winner(x)
    #place the colored markers 
    plt.plot(w[0]+0.5,w[1]+0.5,markers[target[i]],markerfacecolor="None",markeredgecolor=colors[target[i]],markersize=12,markeredgewidth=2)
```

```python
# putting it all together 
plt.bone()
plt.pcolor(som.distance_map())
plt.colorbar()
#using markers for visualising
markers=['o','s','D']
colors=['r','g','b']# red :setosa g: versicolor blue diamond verginica

for i,x in enumerate(data):
    #get the winning nodes
    w=som.winner(x)
    #place the colored markers 
    plt.plot(w[0]+0.5,w[1]+0.5,markers[target[i]],markerfacecolor="None",markeredgecolor=colors[target[i]],markersize=12,markeredgewidth=2)
```
